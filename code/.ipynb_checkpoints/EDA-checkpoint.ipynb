```python
categorical_columns = ['driver_relationship', 'incident_type', 'collision_type','incident_severity','authorities_contacted','policy_state','customer_gender']  # Replace with your actual column names

# Read the CSV file and explicitly set the data types of categorical columns as 'category' or elese pandas treat each cataegory as a new column
df_raw = pd.read_csv(r'E:\AWS Machine learning\full pipeline project\original data\FOR_EDA.csv', dtype={col: 'category' for col in categorical_columns})
df_raw.dtypes
```

```python
# Number of observations and features
df_raw.shape
```

```python
# This clearly shows California has hoghest fraud cases
fig = plt.figure(figsize=(10,6))
ax = df_raw.groupby('policy_state').fraud.count().plot.bar(ylim=0)
ax.set_ylabel('fraud')
plt.show()
```

```python
# Total claim amounts in each month of the year
# January,October and November has highest claims compared to other months
plt.style.use('fivethirtyeight')
fig = plt.figure(figsize=(18,6))
ax = df_raw.groupby('incident_month').total_claim_amount.count().plot.bar(ylim=0)
ax.set_ylabel('Claim amount ($)')
plt.show()
```

```python
# Vehicles claimed to be met with collision is the most fraud cases and least being the theft cases.

plt.style.use('fivethirtyeight')
fig = plt.figure(figsize=(10,6))
ax = df_raw.groupby('incident_type').fraud.count().plot.bar(ylim=0)
ax.set_xticklabels(ax.get_xticklabels(), rotation=20, ha="right")
ax.set_ylabel('Fraud reported')
plt.show()
```

```python
# Incident severity category belonging to Minority had most fraud cases, second being Major and least being Totaled. 
plt.style.use('fivethirtyeight')
fig = plt.figure(figsize=(10,6))
ax = df_raw.groupby('incident_severity').fraud.count().plot.bar(ylim=0)
ax.set_xticklabels(ax.get_xticklabels(), rotation=20, ha="right")
ax.set_ylabel('Fraud reported')
plt.show()
```

```python
# Insurance claims with driver relationship as self has the highest fradulent cases.

plt.style.use('fivethirtyeight')
fig = plt.figure(figsize=(10,6))
ax = df_raw.groupby('driver_relationship').fraud.count().plot.bar(ylim=0)
ax.set_xticklabels(ax.get_xticklabels(), rotation=20, ha="right")
ax.set_ylabel('Fraud reported')
plt.show()
```

```python
#Fraud cases by insured relationship 

import matplotlib.pyplot as plt

plt.style.use('fivethirtyeight')


fraud_counts_relationship = df_raw['fraud'].groupby(df_raw['driver_relationship']).sum()

# Calculate the percentage of each insured relationship
total_fraud_cases_relationship = fraud_counts_relationship.sum()
percentages_relationship = [(count / total_fraud_cases_relationship) * 100 for count in fraud_counts_relationship]

# Create a pie chart
fig, ax = plt.subplots(figsize=(10, 6))
colors_relationship = ['#008fd5', '#fc4f30', '#e5ae38', 'purple', 'lightblue'] 
wedges_relationship, _ = ax.pie(fraud_counts_relationship, labels=None, autopct=None, startangle=90, colors=colors_relationship)
ax.set_aspect('equal')  # Equal aspect ratio ensures that the pie chart is circular
ax.set_title('Fraud Cases by Insured Relationship', fontsize=16)

# Create the legend
ax.legend(wedges_relationship, [f'{label} ({percentage:.1f}%)' for label, percentage in zip(fraud_counts_relationship.index, percentages_relationship)], 
          title='Insured Relationship', fontsize=12, loc='best')

plt.show()

```

```python
# Authorities contacted when fraud claims were made is Police, which is quite surprising. As this should not be the fraud case.

plt.style.use('fivethirtyeight')
fig = plt.figure(figsize=(10,6))
ax = df_raw.groupby('authorities_contacted').fraud.count().plot.bar(ylim=0)
ax.set_xticklabels(ax.get_xticklabels(), rotation=20, ha="right")
ax.set_ylabel('Fraud reported')
plt.show()
```

```python
# with 0 being the number of witness when farud cases were filled is maximum which is as expected. 

plt.style.use('fivethirtyeight')
fig = plt.figure(figsize=(10,6))
ax = df_raw.groupby('num_witnesses').fraud.count().plot.bar(ylim=0)
ax.set_xticklabels(ax.get_xticklabels(), rotation=20, ha="right")
ax.set_ylabel('Fraud reported')
plt.show()
```

```python
# Number of fradulent cases are considerably high between 8am to 4pm. Least at 9pm-3am. 

plt.style.use('fivethirtyeight')
fig = plt.figure(figsize=(10,6))
ax = df_raw.groupby('incident_hour').fraud.count().plot.bar(ylim=0)
ax.set_xticklabels(ax.get_xticklabels(), rotation=20, ha="right")
ax.set_ylabel('Fraud reported')
plt.show()
```

```python
#People with Bachelors degree do a maximum fradulent claims and people with education below high school do less fraud claims.

plt.style.use('fivethirtyeight')
fig = plt.figure(figsize=(10,6))
ax = df_raw.groupby('customer_education').fraud.count().plot.bar(ylim=0)
ax.set_xticklabels(ax.get_xticklabels(), rotation=25, ha="right")
ax.set_ylabel('Fraud reported')
plt.show()
```

```python
# Both male and female approximately claim equal fraud claims

plt.style.use('fivethirtyeight')
fig = plt.figure(figsize=(10,6))
ax = df_raw.groupby('customer_gender').fraud.count().plot.bar(ylim=0)
ax.set_xticklabels(ax.get_xticklabels(), rotation=25, ha="right")
ax.set_ylabel('Fraud reported')
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.style.use('fivethirtyeight')


fraud_counts = df_raw['fraud'].groupby(df_raw['customer_gender']).sum()

# Calculate the percentage of each gender
total_fraud_cases = fraud_counts.sum()
percentages = [(count / total_fraud_cases) * 100 for count in fraud_counts]


fig, ax = plt.subplots(figsize=(10, 7))
colors = ['#008fd5', '#fc4f30', '#e5ae38', 'purple'] 
wedges, _ = ax.pie(fraud_counts, labels=None, autopct=None, startangle=90, colors=colors)
ax.set_aspect('equal')  # Equal aspect ratio ensures that the pie chart is circular


ax.set_title('Fraud Cases by Customer Gender', fontsize=16)

# legend with percentages
ax.legend(wedges, [f'{label} ({percentage:.1f}%)' for label, percentage in zip(fraud_counts.index, percentages)], 
          title='Customer Gender', fontsize=12, loc='best')

plt.show()

```

```python
# This describes that Annual Policy with highest Premium gets more fraud claims. 
import matplotlib.pyplot as plt

plt.style.use('fivethirtyeight')


fraud_counts = df_raw['fraud'].groupby(df_raw['policy_annual_premium']).sum()

# Calculate the percentage of each gender
total_fraud_cases = fraud_counts.sum()
percentages = [(count / total_fraud_cases) * 100 for count in fraud_counts]


fig, ax = plt.subplots(figsize=(10, 7))
colors = ['#008fd5', '#fc4f30', '#e5ae38', 'purple'] 
wedges, _ = ax.pie(fraud_counts, labels=None, autopct=None, startangle=90, colors=colors)
ax.set_aspect('equal')  # Equal aspect ratio ensures that the pie chart is circular


ax.set_title('Fraud Cases by policy_annual_premium', fontsize=16)

# legend with percentages
ax.legend(wedges, [f'{label} ({percentage:.1f}%)' for label, percentage in zip(fraud_counts.index, percentages)], 
          title='policy_annual_premium', fontsize=12, loc='best')

plt.show()

```
